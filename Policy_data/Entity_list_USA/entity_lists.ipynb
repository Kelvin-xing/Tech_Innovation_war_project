# Loading Specially Designated Nationals List
From https://ofac.treasury.gov/specially-designated-nationals-list-data-formats-data-schemas

```python

### Download zip of xml and xsd
import requests
import zipfile
import io
url_zip = 'https://www.treasury.gov/ofac/downloads/sanctions/1.0/sdn_advanced.zip'
response = requests.get(url_zip)

## Extracting contents
with zipfile.ZipFile(io.BytesIO(response.content)) as z:
    z.extractall('Tech_Innovation_war_project/SDN')
```

```python
### Open list
import pandas as pd
from pprint import pprint
import xml.etree.ElementTree as ET
import xml.sax as SX
import numpy as np

with open('/Users/mingjiexing/Desktop/Python/Tech_Innovation_war_project/Policy_data/SDN/sdn_advanced.xml','r') as xml_file:
    root = SX.parse(xml_file)
tags = [elem.tag for elem in root.iter()]
tags = np.unique(tags)
print(tags)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# Load the XML document
tree = ET.parse('../SDN/sdn_advanced.xml')
root = tree.getroot()

# Create an empty dataframe
df = pd.DataFrame(columns=['NodeName', 'NodeValue'])

# Get the first 10 nodes
for i, child in enumerate(root):
    if i < 10:
        node_name = child.tag
        node_name = node_name.replace("{http://www.un.org/sanctions/1.0}", "")

        node_value = ET.tostring(child.attrib).decode('utf-8')
        df = df.append({'NodeName': node_name, 'NodeValue': node_value}, ignore_index=True)
    else:
        break

# Print the dataframe
print(df)
```

```python
import os
import time
import xml.etree.cElementTree as ET

def get_file_size_in_mb(file_name):
    file_size = 0
    if os.path.isfile(file_name):
        file_size = os.path.getsize(file_name)
        file_size = round(file_size / (1024*1024.0), 2)
    file_size = '{:,.2f}'.format(file_size)
    return str(file_size)+'MB'

def read_xml_file(xml_file, element):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    number_of_element = len(root.findall(element))
    return '{:,.0f}'.format(number_of_element)

if __name__ == '__main__':
    xml_file = '../SDN/sdn_advanced.xml'
    print(f'File Size: {get_file_size_in_mb(xml_file)}')
    start_time = time.perf_counter()
    counter = read_xml_file(xml_file, 'Sanctions/SanctionsEntry')
    end_time = time.perf_counter()
    total_time = round(end_time-start_time,2)
    print(f'xml.etree.cElementTree - Total time taken:[{total_time}] seconds'
          f'to identify the number of elements:[{counter}]')
```
